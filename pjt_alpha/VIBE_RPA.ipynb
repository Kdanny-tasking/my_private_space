```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
import pandas as pd
import time

# 1. 크롬 브라우저 열기 (headless=False)
options = webdriver.ChromeOptions()
options.headless = False  # 반드시 False

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# 결과 저장 리스트
data_list = []

# 5. 1page ~ 9page 반복
for page in range(1, 10):
    url = f"https://finance.naver.com/news/mainnews.naver?&page={page}"
    driver.get(url)
    
    time.sleep(2)  # 로딩 대기
    
    # 3. class 이름이 'articleSubject'인 요소 가져오기
    articles = driver.find_elements(By.CLASS_NAME, "articleSubject")
    
    # 4. 반복문으로 제목 + URL 출력
    for article in articles:
        title = article.text
        
        link = article.find_element(By.TAG_NAME, "a").get_attribute("href")
        
        print(f"[{page}] {title}")
        print(link)
        print("-" * 50)
        
        # 데이터 저장
        data_list.append([page, title, link])

# 브라우저 종료
driver.quit()
```

```python
# DataFrame 생성
df = pd.DataFrame(data_list, columns=["page", "뉴스제목", "뉴스URL"])

# 6. 엑셀 저장 (인코딩 설정 X)
df.to_excel("naver_finance_news.xlsx", index=False)

print("엑셀 저장 완료!")
```

#### 2-1-3-2 위의 기능을 반복문 100번으로 확장하기

```python
from pathlib import Path  # 파일/폴더 경로를 쉽게 다루기 위한 모듈

# 작업할 대상 폴더 경로 설정 (현재 폴더 기준)
folder_path = Path("엑셀파일정리")

# glob("excel*.xlsx")
# → 'excel'로 시작하고 '.xlsx'로 끝나는 모든 파일을 가져옴
# 예: excel1.xlsx, excel23.xlsx 등
for file in folder_path.glob("excel*.xlsx"):
    
    # file.stem
    # → 파일명에서 확장자를 제외한 부분
    # 예: "excel12.xlsx" → "excel12"
    filename_without_ext = file.stem
    
    # "excel" 문자열 제거 → 숫자 부분만 추출
    # 예: "excel12" → "12"
    num_str = filename_without_ext.replace("excel", "")
    
    # 숫자인 경우만 처리 (혹시 이상한 파일명 필터링)
    if num_str.isdigit():
        
        # int(num_str)
        # → 문자열 숫자를 정수로 변환 (예: "01" → 1)
        
        # :03d
        # → 숫자를 3자리로 맞추고, 부족하면 앞을 0으로 채움
        # 예:
        # 1   → 001
        # 12  → 012
        # 100 → 100
        formatted_num = f"{int(num_str):03d}"
        
        # 새로운 파일 이름 생성
        # 예: "excel001.xlsx"
        new_name = f"excel{formatted_num}.xlsx"
        
        # 기존 경로 + 새 파일명으로 새로운 경로 생성
        new_file = folder_path / new_name
        
        # 파일 이름 변경 (실제 작업 수행)
        file.rename(new_file)

# 모든 작업 완료 메시지 출력
print("파일 이름 일괄 변경 완료")
```

### 2-1-4. 프로세스4

#### 2-1-4-1. 현재폴더의 하위에 "엑셀파일정리" 라는 폴더의 파일을 25개씩 묶은 수 만큼 폴더를 하위 폴더 생성을 해줘

```python
from pathlib import Path
import math  # 올림 계산을 위해 사용

# 대상 폴더
folder_path = Path("엑셀파일정리")

# 엑셀 파일 목록 가져오기
# excel로 시작하고 .xlsx로 끝나는 파일들
files = list(folder_path.glob("excel*.xlsx"))

# 전체 파일 개수
file_count = len(files)

# 한 폴더당 파일 개수
group_size = 25

# 필요한 폴더 개수 계산 (올림 처리)
# 예: 100개 → 4개 / 101개 → 5개
num_groups = math.ceil(file_count / group_size)

# 폴더 생성
for i in range(1, num_groups + 1):
    
    # group1, group2, group3 ... 형태
    subfolder = folder_path / f"group{i}"
    
    # 폴더 생성 (이미 있으면 무시)
    subfolder.mkdir(exist_ok=True)

print(f"{num_groups}개의 하위 폴더 생성 완료")
```

#### 2-1-4-1. 해당 하위 폴더에 25개씩 나눠서 넣어줘

```python
from pathlib import Path
import math

# 1. 대상 폴더 설정
folder_path = Path("엑셀파일정리")

# 2. 엑셀 파일 목록 가져오기 + 정렬
# 정렬을 해야 순서대로 25개씩 정확히 나눠짐
files = sorted(folder_path.glob("excel*.xlsx"))

# 3. 한 그룹당 파일 개수
group_size = 25

# 4. 필요한 폴더 개수 계산 (올림)
num_groups = math.ceil(len(files) / group_size)

# 5. 폴더 생성 + 파일 이동
for i in range(num_groups):
    
    # group1, group2 ...
    subfolder = folder_path / f"group{i+1}"
    subfolder.mkdir(exist_ok=True)
    
    # 현재 그룹에 들어갈 파일 범위 계산
    start_index = i * group_size
    end_index = start_index + group_size
    
    # 해당 범위의 파일들만 선택
    group_files = files[start_index:end_index]
    
    # 파일 이동
    for file in group_files:
        destination = subfolder / file.name
        
        # rename() → 이동 + 이름 유지
        file.rename(destination)

print("파일을 25개씩 폴더에 나눠서 이동 완료")
```

```python
from pathlib import Path  # 파일/폴더 경로를 객체처럼 다루기 위한 모듈
import math               # 올림 계산을 위해 사용

# --------------------------------------------------
# 1. 기준 폴더 설정
# --------------------------------------------------
# 현재 실행 위치 기준으로 "엑셀파일정리" 폴더를 지정
folder_path = Path("엑셀파일정리")

# --------------------------------------------------
# 2. 대상 파일 목록 수집
# --------------------------------------------------
# glob("excel*.xlsx")
# → "excel"로 시작하고 ".xlsx"로 끝나는 모든 파일 검색
# 예: excel001.xlsx, excel025.xlsx ...
#
# sorted()를 사용하는 이유:
# → 파일을 순서대로 정렬하지 않으면
#   그룹이 뒤죽박죽 섞일 수 있음 (매우 중요)
files = sorted(folder_path.glob("excel*.xlsx"))

# --------------------------------------------------
# 3. 기본 설정값 정의
# --------------------------------------------------
group_size = 25  # 한 폴더에 넣을 파일 개수

# 전체 파일 개수
total_files = len(files)

# 필요한 폴더 개수 계산
# math.ceil() → 나눴을 때 올림 처리
# 예:
# 100개 → 4개
# 101개 → 5개
num_groups = math.ceil(total_files / group_size)

# --------------------------------------------------
# 4. 그룹별 폴더 생성 + 파일 이동
# --------------------------------------------------
for i in range(num_groups):
    
    # -------------------------------
    # 4-1. 그룹 폴더 생성
    # -------------------------------
    # i는 0부터 시작하므로 +1 해서 group1, group2 ... 형태로 만듦
    subfolder = folder_path / f"group{i + 1}"
    
    # mkdir(exist_ok=True)
    # → 이미 폴더가 존재해도 에러 발생하지 않도록 처리
    subfolder.mkdir(exist_ok=True)
    
    # -------------------------------
    # 4-2. 현재 그룹에 해당하는 파일 범위 계산
    # -------------------------------
    # 예: group1 → 0~24
    #     group2 → 25~49
    start_index = i * group_size
    end_index = start_index + group_size
    
    # 리스트 슬라이싱으로 해당 구간 파일 추출
    group_files = files[start_index:end_index]
    
    # -------------------------------
    # 4-3. 파일 이동 처리
    # -------------------------------
    for file in group_files:
        
        # 이동할 목적지 경로 생성
        # (파일명은 그대로 유지)
        destination = subfolder / file.name
        
        # rename()
        # → 파일 이름 변경 + 경로 이동을 동시에 수행
        # → 여기서는 "이동" 기능으로 사용됨
        file.rename(destination)

# --------------------------------------------------
# 5. 작업 완료 안내
# --------------------------------------------------
print(f"총 {total_files}개 파일을 {num_groups}개 폴더로 나누어 이동 완료")
```

```python
import os
import shutil

# 대상 폴더 경로
base_path = r"C:\Users\student\Desktop\Vibe_python\파일자동분류기"

# 파일 목록 가져오기
files = [f for f in os.listdir(base_path) if os.path.isfile(os.path.join(base_path, f))]

for file in files:
    try:
        # 파일명 분리
        name, ext = file.rsplit('.', 1)
        dept, doc_type, year, num = name.split('_')

        # 이동할 폴더 경로 (부서/문서종류)
        target_dir = os.path.join(base_path, dept, doc_type)

        # 폴더 없으면 생성
        os.makedirs(target_dir, exist_ok=True)

        # 원본 파일 경로
        src = os.path.join(base_path, file)

        # 이동될 경로
        dst = os.path.join(target_dir, file)

        # 파일 이동
        shutil.move(src, dst)

        print(f"이동 완료: {file} → {dept}/{doc_type}")

    except Exception as e:
        print(f"에러 발생 ({file}): {e}")
```

```python
import os        # 파일/폴더 경로 및 목록을 다루기 위한 기본 라이브러리
import shutil    # 파일 이동, 복사 등을 위한 라이브러리

# 📌 정리할 대상 폴더 경로
# r"" 를 쓰면 \ (역슬래시)를 그대로 문자열로 사용할 수 있음 (윈도우 필수)
base_path = r"C:\Users\student\Desktop\Vibe_python\파일자동분류기"


# 📌 해당 폴더 안의 "파일"만 리스트로 가져오기
# os.listdir() → 폴더 안 모든 요소(파일 + 폴더) 반환
# os.path.isfile() → 파일인지 확인 (폴더는 제외)
files = [
    f for f in os.listdir(base_path)
    if os.path.isfile(os.path.join(base_path, f))
]


# 📌 파일 하나씩 반복 처리
for file in files:
    try:
        # -------------------------------
        # 1️⃣ 파일명과 확장자 분리
        # -------------------------------
        # rsplit('.', 1)
        # → 오른쪽에서부터 '.' 기준으로 1번만 나눔
        # 예: "가전_기획안_2022_54.docx"
        # → name = "가전_기획안_2022_54"
        # → ext  = "docx"
        name, ext = file.rsplit('.', 1)


        # -------------------------------
        # 2️⃣ 파일명 구조 분해
        # -------------------------------
        # "_" 기준으로 나누면 4개 요소가 나와야 정상
        # [부서, 문서종류, 연도, 번호]
        dept, doc_type, year, num = name.split('_')


        # -------------------------------
        # 3️⃣ 이동할 폴더 경로 생성
        # -------------------------------
        # 예: C:\...\파일자동분류기\가전\기획안
        target_dir = os.path.join(base_path, dept, doc_type)


        # -------------------------------
        # 4️⃣ 폴더 없으면 자동 생성
        # -------------------------------
        # exist_ok=True
        # → 이미 폴더가 있어도 에러 안 나게 처리
        os.makedirs(target_dir, exist_ok=True)


        # -------------------------------
        # 5️⃣ 원본 파일 경로 생성
        # -------------------------------
        src = os.path.join(base_path, file)


        # -------------------------------
        # 6️⃣ 이동될 최종 경로 생성
        # -------------------------------
        dst = os.path.join(target_dir, file)


        # -------------------------------
        # 7️⃣ 파일 이동
        # -------------------------------
        # shutil.move()
        # → 파일을 잘라내기(Cut)해서 다른 폴더로 이동
        shutil.move(src, dst)


        # -------------------------------
        # 8️⃣ 로그 출력 (진행 상황 확인용)
        # -------------------------------
        print(f"이동 완료: {file} → {dept}/{doc_type}")


    except Exception as e:
        # -------------------------------
        # ⚠️ 예외 처리
        # -------------------------------
        # 파일명이 규칙과 다르거나 오류 발생 시
        # 프로그램이 멈추지 않고 다음 파일 계속 처리
        print(f"에러 발생 ({file}): {e}")
    
```

## 2-3. 모든 코드를 질문만으로 만들 수 있을까?
- 한국서부발전의 홈페이지에 들어가서,
- 보도자료를 1~10 페이지까지 모두 가져와서
- 엑셀로 제작해주는 파이썬 코드 만들어줘.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

# 결과 저장 리스트
data = []

# 기본 URL (보도자료 게시판)
base_url = "https://www.iwest.co.kr/cop/bbs/selectBoardList.do"

# 파라미터 (사이트 구조에 따라 바뀔 수 있음)
params = {
    "bbsId": "BBSMSTR_000000000138",  # 보도자료 게시판 ID
    "pageIndex": 1
}

for page in range(1, 11):  # 1~10 페이지
    print(f"{page} 페이지 수집 중...")
    
    params["pageIndex"] = page
    
    response = requests.get(base_url, params=params)
    response.raise_for_status()
    
    soup = BeautifulSoup(response.text, "html.parser")
    
    # 게시글 테이블 찾기 (사이트 구조에 따라 수정 필요)
    rows = soup.select("table tbody tr")
    
    for row in rows:
        cols = row.find_all("td")
        
        if len(cols) < 4:
            continue
        
        title = cols[1].get_text(strip=True)
        date = cols[3].get_text(strip=True)
        
        # 링크 가져오기
        link_tag = cols[1].find("a")
        if link_tag:
            link = "https://www.iwest.co.kr" + link_tag.get("href")
        else:
            link = ""
        
        data.append({
            "제목": title,
            "날짜": date,
            "링크": link
        })
    
    time.sleep(1)  # 서버 부담 방지

# 데이터프레임 생성
df = pd.DataFrame(data)

# 엑셀 저장
file_name = "한국서부발전_보도자료.xlsx"
df.to_excel(file_name, index=False)

print(f"완료! {file_name} 파일이 생성되었습니다.")
```

```python
import time
st = 3

# 1) 인터넷 브라우저 열기 (AI가 어려워함)
#!pip install selenium
from selenium import webdriver
driver = webdriver.Edge("c:/msedgedriver.exe")

time.sleep(st)

# 2) 원하는 사이트 접속하기
driver.get("https://www.naver.com")

time.sleep(st)

# 3) 찾고자 하는 웹요소의 위치 찾기 (AI가 어려워함)
# 미션 : 초록색 검색창 웹요소를 찾아서 green_btn_ele 에 담기
from selenium.webdriver.common.by import By
# green_btn_ele = driver.find_element(By.??, "~~")
green_btn_ele = driver.find_element(By.ID, "query")

time.sleep(st)

# 4) 웹요소에 데이터 입력하기
green_btn_ele.send_keys("반도체")

time.sleep(st)

# 5-1) 웹요소를 클릭하기
# 미션 : 검색돋보기 버튼을 gr_se_btn_ele 에 담아서 클릭하기
# 조건 : ID로 찾지말기
# 힌트 : ctrl + F를 누르면 검색가능
# 힌트2 : 부모를 데리고 오면 자식이 따라옴
gr_se_btn_ele = driver.find_element(By.CLASS_NAME, "ico_btn_search_svg")
gr_se_btn_ele.click()

time.sleep(st)

# 5-2) "카페" 탭을 클릭하기
# 프롬프트 : class 이름이 'tab'인 모든 웹 요소를 찾고
# 그 중에서 "카페"라는 텍스트를 가진 웹 요소에 있는 텍스트를 출력하고
# 클릭해줘
# cafe_btn_ele = driver.find_element(By.LINK_TEXT, "카페")
from selenium.webdriver.common.by import By

time.sleep(st)

# 1. 'tab'이라는 class를 가진 모든 요소를 리스트 형태로 가져옵니다 
tab_elements = driver.find_elements(By.CLASS_NAME, "tab")

time.sleep(st)

# 2. 리스트를 순회하며 텍스트가 "카페"인 요소를 찾습니다
for tab in tab_elements:
    if tab.text == "카페":
        print(f"찾은 요소의 텍스트: {tab.text}") # 텍스트 출력
        tab.click()                             # 해당 요소 클릭
        break                                   # 원하는 요소를 찾았으므로 반복문 종료

time.sleep(st)

# 6) 글 출력
from selenium.webdriver.common.by import By

# 'title_link' 클래스를 가진 모든 요소를 가져옵니다
elements = driver.find_elements(By.CLASS_NAME, "title_link")

if not elements:
    print("해당 클래스를 가진 게시글을 찾을 수 없습니다.")
else:
    for index, item in enumerate(elements, 1):
        try:
            # 1. 텍스트 추출 (요소 자체의 텍스트)
            text = item.text
            
            # 2. URL 추출 (요소 자체가 <a> 태그이므로 바로 href 속성을 가져옵니다)
            url = item.get_attribute("href")
            
            # 3. 결과 출력
            print(f"[{index}번째 게시글]")
            print(f"텍스트: {text}")
            print(f"URL: {url}")
            print("-" * 40)
            
        except Exception as e:
            # 혹시 모를 다른 에러를 확인하기 위해 에러 내용(e)을 함께 출력합니다
            print(f"[{index}번째 게시글] 처리 중 에러 발생: {e}")
            print("-" * 40)

time.sleep(st)
```

# 발전
- 위의 코드를 기반으로 아래의 기능을 추가하여 발전시켜줘
- 1) "반도체"말고 임의의 투자섹터로 5개의 단어가 조사되게해줘.
  2) #1의 동작은 한번만 하고, #2~#6의 동작이 반복되게 해줘.

```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By

# 1) 조사할 투자 섹터 리스트 정의
keywords = ["반도체", "이차전지", "AI", "로봇", "바이오"]
st = 2 # 대기 시간 조정

# 브라우저 초기화 (최초 1회만 실행)
driver = webdriver.Edge("c:/msedgedriver.exe")
driver.get("https://www.naver.com")
time.sleep(st)

# 2~6) 반복문 시작
for word in keywords:
    print(f"\n>>> '{word}' 섹터 조사 시작")
    
    # 2) 검색창 찾기 및 입력
    query_box = driver.find_element(By.ID, "query")
    query_box.clear() # 이전 검색어 삭제
    query_box.send_keys(word)
    
    # 검색 버튼 클릭 (이전 코드 방식)
    search_btn = driver.find_element(By.CLASS_NAME, "ico_btn_search_svg")
    search_btn.click()
    time.sleep(st)
    
    # 3) '카페' 탭 찾아서 클릭
    tab_elements = driver.find_elements(By.CLASS_NAME, "tab")
    for tab in tab_elements:
        if tab.text == "카페":
            tab.click()
            time.sleep(st)
            break
            
    # 4) 게시글 정보 추출
    elements = driver.find_elements(By.CLASS_NAME, "title_link")
    
    # 상위 3개 정도만 출력하도록 제한 (너무 길어지는 것 방지)
    for index, item in enumerate(elements[:3], 1):
        text = item.text
        url = item.get_attribute("href")
        print(f"[{index}] {text}")
        print(f"URL: {url}")
        
    # 다시 검색창으로 돌아가기 위해 네이버 메인으로 이동
    driver.get("https://www.naver.com")
    time.sleep(st)

# 작업 종료
driver.quit()
print("\n모든 조사가 완료되었습니다.")
```

```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By

# 1) 설정
keywords = ["반도체", "이차전지", "AI", "로봇", "바이오"]
st = 3 # 로딩 시간을 위해 3초로 약간 늘림
all_data = [] 

driver = webdriver.Edge("c:/msedgedriver.exe")

for word in keywords:
    driver.get("https://www.naver.com")
    time.sleep(st)
    
    # 검색 수행
    query_box = driver.find_element(By.ID, "query")
    query_box.send_keys(word)
    search_btn = driver.find_element(By.CLASS_NAME, "ico_btn_search_svg")
    search_btn.click()
    time.sleep(st)
    
    # 카페 탭 클릭
    tab_elements = driver.find_elements(By.CLASS_NAME, "tab")
    for tab in tab_elements:
        if tab.text == "카페":
            tab.click()
            time.sleep(st)
            break
            
    # 게시글 정보 추출
    elements = driver.find_elements(By.CLASS_NAME, "title_link")
    
    # 데이터 누락 방지를 위해 루프 안에서 확인
    for item in elements[:3]:
        text = item.text
        url = item.get_attribute("href")
        # 여기서 확실하게 [키워드, 제목, URL] 구조로 삽입
        all_data.append([word, text, url])
        print(f"추가됨: {word} | {text}") # 터미널에서 데이터 확인용

driver.quit()

# 엑셀 저장
df = pd.DataFrame(all_data, columns=["키워드", "제목", "url"])
df.to_excel("nav_cafe_list.xlsx", index=False)

print("\n완료: 'nav_cafe_list.xlsx' 파일이 생성되었습니다.")
```

```python
import pandas as pd
import os
import win32com.client as win32

# 1. 현재 작업 경로를 기준으로 파일의 절대 경로 설정
file_name = "상대현재경로.xlsx"
file_path = os.path.abspath(file_name)

# 2. win32com을 사용하여 엑셀 실행
excel = win32.gencache.EnsureDispatch('Excel.Application')
excel.Visible = False  # 엑셀 화면을 띄우지 않음

# 3. 엑셀 파일 열기 (읽기 전용 모드)
wb = excel.Workbooks.Open(file_path)
ws = wb.Sheets(1)  # 첫 번째 시트 선택

# 4. 전체 데이터 범위 가져오기
data = ws.UsedRange.Value

# 5. Pandas DataFrame으로 변환
# 데이터의 첫 번째 행을 헤더(열 이름)로 사용
df3 = pd.DataFrame(list(data[1:]), columns=data[0])

# 6. 엑셀 종료
wb.Close(False)
excel.Quit()

# 결과 확인
print(df3)
```
