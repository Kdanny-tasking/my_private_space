# 최종프로젝트

```python
st = 3
import time
# 1) 인터넷 자동화
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By

# 1) '반도체'를 제외한 5개의 투자 섹터 정의
keywords = ["이차전지", "AI", "로봇", "바이오", "클라우드"]
st = 2
all_data = [] 

# 브라우저 초기화
driver = webdriver.Edge("c:/msedgedriver.exe")

for word in keywords:
    driver.get("https://www.naver.com")
    time.sleep(st)
    
    # 검색 수행
    query_box = driver.find_element(By.ID, "query")
    query_box.send_keys(word)
    search_btn = driver.find_element(By.CLASS_NAME, "ico_btn_search_svg")
    search_btn.click()
    time.sleep(st)
    
    # 카페 탭 클릭
    tab_elements = driver.find_elements(By.CLASS_NAME, "tab")
    for tab in tab_elements:
        if tab.text == "카페":
            tab.click()
            time.sleep(st)
            break
            
    # 게시글 정보 추출
    elements = driver.find_elements(By.CLASS_NAME, "title_link")
    
    # 각 키워드당 3개씩 추출하여 [키워드, 제목, URL] 리스트에 저장
    for item in elements[:3]:
        text = item.text
        url = item.get_attribute("href")
        all_data.append([word, text, url])
        print(f"데이터 수집 완료: {word} - {text}") 

driver.quit()

# 엑셀 저장
df = pd.DataFrame(all_data, columns=["키워드", "제목", "url"])
df.to_excel("nav_cafe_list.xlsx", index=False)

print("\n완료: 'nav_cafe_list.xlsx' 파일이 생성되었습니다.")
time.sleep(st)



# 2) 엑셀 자동화
import pandas as pd
df = pd.read_excel("nav_cafe_list.xlsx")

# 1. 키워드별로 그룹화하여 개수(size)를 구하고 데이터프레임으로 변환
keyword_count = df.groupby('키워드').size().reset_index(name='게시글_수')

# 2. 엑셀 파일로 저장
file_name = 'DE.xlsx'

# with 구문을 사용하면 파일을 안전하게 생성하고 닫을 수 있습니다.
with pd.ExcelWriter(file_name, engine='openpyxl') as writer:
    # '키워드별_게시글_요약'이라는 시트 이름으로 저장
    keyword_count.to_excel(writer, index=False, sheet_name='키워드별_게시글_요약')

print(f"'{file_name}' 파일이 성공적으로 저장되었습니다.")
time.sleep(st)




# 3) 이메일 자동화
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.base import MIMEBase
from email.mime.text import MIMEText
from email import encoders

# 1. 이메일 설정 정보
smtp_server = "smtp.gmail.com"
smtp_port = 587
sender_email = "danny.tasking@gmail.com"       # 보내는 사람 이메일
sender_password = "ttqukiwfqmzljgng"      # 구글 앱 비밀번호 (일반 비밀번호 아님)
receiver_email = "rlawofyd88@gmail.com" # 받는 사람 이메일

# 2. 메시지 객체 생성
msg = MIMEMultipart()
msg['From'] = sender_email
msg['To'] = receiver_email
msg['Subject'] = "자료조사내용 보내드립니다."

# 3. 본문 내용 추가
body = "신기술 기반 네이버 카페 반응 현황을 정리해서 보내드립니다."
msg.attach(MIMEText(body, 'plain'))

# 4. 파일 첨부
file_path = "DE.xlsx"
with open(file_path, "rb") as attachment:
    part = MIMEBase("application", "octet-stream")
    part.set_payload(attachment.read())
    encoders.encode_base64(part)
    part.add_header("Content-Disposition", f"attachment; filename= {file_path}")
    msg.attach(part)

# 5. 이메일 서버 접속 및 전송
try:
    server = smtplib.SMTP(smtp_server, smtp_port)
    server.starttls()  # 보안 시작
    server.login(sender_email, sender_password)
    server.sendmail(sender_email, receiver_email, msg.as_string())
    print("이메일이 성공적으로 전송되었습니다.")
except Exception as e:
    print(f"오류 발생: {e}")
finally:
    server.quit()
time.sleep(st)
```
